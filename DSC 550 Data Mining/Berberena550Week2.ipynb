```python
# To bring the dataset into Python and manipulate the data with the end goal of visualizing multiple graphs and plots, both 
# Pandas and Matplotlib.pyplot need to be imported

import pandas as pd
import matplotlib.pyplot as plt

# The reading of the dataset into a Pandas data frame is performed here using Pandas's read_csv() function and stored into 
# a variable which will used to call the data frame when needed

immigration = pd.read_csv('Immigration_data.csv')

# To see the column headers (variable names), the columns keyword is used. This allows us to see the specific variables 
# within the dataset to explain the purpose of the content.

immigration.columns
```

```python
# Introduction

# With the help of Kaggle (where I retrieved this dataset), I was able to find a clean dataset that outlined American 
# immigration statistics from 1980 to 2021. The only thing I have done already to this dataset before loading it here is 
# change the variable names slightly by shortening a couple of the long variable names and adding underscores for better 
# variable calling conventions. This dataset is fairly small and easy to work with, which works great for the 
# purposes of this assignment. Now the immigration statistics that are listed in the dataset are easily seen with the 
# above code that has been created. Aside from the Year variable that defines the dataset as having observations from 1980 
# to 2021, the remaining variables we see are as follows: 

# 1. Immigrants Obtaining Lawful Permanent Resident Status (Immigrant_Perm_Residence): This variable shows the number of 
#    immigrants who obtained green cards.

# 2. Refugee Arrivals (Refugee_Arrivals): The number of refugees who arrived in the United States is highlighted by this 
#    variable. According to Kaggle, the dataset excludes Amerasian immigrants except for the fiscal years of 1989 and 1991.

# 3. Noncitizen Apprehensions (Noncitizen_Aprehensions): The variable here defines the number of noncitizens apprehended in 
#    the United States. The data for years 2020 and 2021 did take into account the apprehensions made based on health 
#    due to COVID-19.

# 4. Noncitizen Removals (Noncitizen_Removals): This variable describes the number of noncitizens removed from the United 
#    States. Removals for the purpose of this dataset have been defined as the compulsory and confirmed movement of an 
#    inadmissible or deportable noncitizen out of the United States based on an order of removal.

# 5. Noncitizen Returns (Noncitizen_Returns): Here the variable highlights the number of noncitizen returns from the United 
#    States. Returns are considered to be the confirmed movement of an inadmissible or deportable noncitizen out of the 
#    United States not based on an order of removal.

# With these variables present, we can see the movement of immigrants and how those movements were classified. The ability 
# to segregate these statistics by year can help to visualize some intriguing plots.   
```

## 2. Identify a question or question(s) that you would like to explore in your data set.

```python
# Looking at the dataset and its variables, I would like to work towards finding answers for these questions: 

# 1. Is there a certain year or range of years where the number of immigrants obtaining lawful permanent resident status 
#    significantly increases in relation to the other years within the dataset? 

# 2. How does the Noncitizen Removals variable behave with the increase of the Noncitizen Apprehensions variable?

# 3. How does the Noncitizen Returns variable behave with the increase of the Refugee Arrivals variable?
```

## 3. Create at least three graphs that help answer these questions. Make sure your graphs are clearly readable and are labeled appropriately and professionally.

```python
# Before plotting any of the variables, the commas inside the values need to be removed so the data points can be 
# recognized as numerical values. The below for loop targets the columns that are recognized as objects and removes the 
# commas with the str.replace() function. The variables that are recognized as objects are then turned into numeric 
# variables with the pd.to_numeric() function. The head() function is simply to see the changes made by the for loop.

for column in immigration.columns:
    if immigration[column].dtype == 'object':
        immigration[column] = immigration[column].str.replace(',', '')
        immigration[column] = pd.to_numeric(immigration[column], errors='coerce')
        
immigration.head()
```

```python
# First graph that will help to answer the first question

# Using Matplotlib's plot() function creates the line plot of the two variables needed to answer the first question (Year 
# and Immigrant_Perm_Residence). The labeling of both axes is accomplished with the xlabel() and ylabel() functions. To add 
# a title to the plot, the title() function is needed, and the whole plot is visualized with the show() function.

plt.plot(immigration.Year, immigration.Immigrant_Perm_Residence)
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Number of Immigrants Obtaining Permanent Residence By Year')
plt.show()
```

```python
# Second graph that will help to answer the second question

# Using Matplotlib's plot() function creates the line plot of the two variables needed to answer the second question 
# (Noncitizen_Aprehensions and Noncitizen_Removals). The plot is put together the same way as the first plot except for the 
# fact that this comparison is much better realized by plotting a scatter plot with the scatter() function instead of a 
# line plot due to the absence of a seemingly linear relationship.

plt.scatter(immigration.Noncitizen_Apprehensions, immigration.Noncitizen_Removals)
plt.xlabel('Noncitizen Apprehensions')
plt.ylabel('Noncitizen Removals')
plt.title('Noncitizen Removals in Relation to Noncitizen Apprehensions')
plt.show()
```

```python
# Third graph that will help to answer the third question

# Using Matplotlib's plot() function creates the line plot of the two variables needed to answer the third question 
# (Refugee_Arrivals and Noncitizen_Returns). The plot is put together the same way as the first plot except for the 
# fact that this comparison is much better realized by plotting a scatter plot with the scatter() function instead of a 
# line plot due to the absence of a seemingly linear relationship.

plt.scatter(immigration.Refugee_Arrivals, immigration.Noncitizen_Returns)
plt.xlabel('Refugee Arrivals')
plt.ylabel('Noncitizen Returns')
plt.title('Noncitizen Returns in Relation to Refugee Arrivals')
plt.show()
```
