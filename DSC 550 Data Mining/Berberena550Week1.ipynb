# Program Start

## 1. Load the dataset as a Pandas data frame.

```python
# To load any dataset as a Pandas data frame, the Pandas library needs to be imported

import pandas as pd

# Loading the Video Game Sales with Ratings CSV file into Python as a Pandas dataframe requires the pd.read_csv() function, 
# and I have specified the delimiter as "," so the Pandas data frame can work on the CSV file as intended. This has been 
# stored into a variable so it can be called on later for the manipulation of the data frame.

videogames = pd.read_csv('video_game_sales.csv', delimiter=',')
```

## 2. Display the first ten rows of data.

```python
# Displaying the first ten rows of data involves the usage of the head() function, which by default only shows the first 
# five rows, yet can be adjusted by simply including the number of rows needed to visualize inside the function parentheses.

videogames.head(10)
```

## 3. Find the dimensions (number of rows and columns) in the data frame. What do these two numbers represent in the context of the data?

```python
rows, columns = videogames.shape
print('Number of rows:', rows)
print('Number of columns:', columns)
```

```python
# The number of rows in a dataset is equivalent to the number of total observations or data points. The number of columns 
# are the total number of variables in the dataset. These variables are characteristics of the data points that can help to 
# classify, categorize, or group them together.
```

## 4. Find the top five games by critic score.

```python
# To sort the dataset by critic score, the sort_values() function is needed. For the data to be sorted by the Critic_Score 
# variable, the by argument is set to that variable. For the data to be displayed with the highest critic score as the 
# first observation, the ascending argument is set to False. The head() function is used to call only the first five data 
# points, which are stored in a variable and then called to be printed.

videogames_by_critic_score = videogames.sort_values(by='Critic_Score', ascending=False)

top_five_critic_score = videogames_by_critic_score.head(5)
top_five_critic_score
```

## 5. Find the number of video games in the data frame in each genre.

```python
# To find out the total number of video games within each genre, the Pandas value_counts() function is needed. Once the 
# Genre variable is called, the value_counts() function adds all of the video games in each specific genre as a Pandas 
# series and that series is stored in a variable to be printed.

genre_counts = videogames['Genre'].value_counts()

print("Number of video games in each genre:")
print(genre_counts)
```

## 6. Find the first five games in the data frame on the SNES platform.

```python
# This scenario asks for the first occurring games in the dataset that are on the SNES platform, so this simply is about 
# setting the Platform variable to SNES and calling it from the dataset to store this new dataset consisting only of SNES 
# games inside another variable. Then the head() function (which defaults to the first five data points) is performed on 
# that variable.

snes_games = videogames[videogames['Platform'] == 'SNES']
snes_games.head()
```

## 7. Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for each publisher to do this.

```python
# To calculate the total global sales by publisher, the groupby() function in Pandas is used to group the data using the 
# Publisher variable. Then the Global_Sales variable is called and the sum() function is applied to that variable, which 
# adds up global sales for each publisher. This line of code in stored in a variable. To display the five publishers with 
# the highest total global sales, the publisher_sum_of_sales variable is sorted by descending order with the use of the 
# sort_values() function with the ascending argument set to False, and the head() function is used to automatically list 
# the top five publishers.

publisher_sum_of_sales = videogames.groupby('Publisher')['Global_Sales'].sum()
publisher_sum_of_sales.sort_values(ascending=False).head()
```

## 8. Create a new column in the data frame that calculates the percentage of global sales from North America. Display the first five rows of the new data frame.

```python
# To create a new column in the data frame for the percentage of North America sales in relation to global sales, I just 
# called a new column name that I made up and set that column equal to the division of NA_Sales and Global_Sales multiplied 
# by 100 as the column will display the values in percent form. The column is automatically appended to the end of the 
# existing dataset and is seen when the head() function is called.

videogames['NA_Sales_Percentage'] = (videogames['NA_Sales'] / videogames['Global_Sales']) * 100
videogames.head()
```

## 9. Find the number NaN entries (missing data values) in each column.

```python
# To find the total NaN entries for each column in the dataset, the isna() function is combined with the sum() function. 
# The isna() function iterates through each column to find the NaN values that are present, and the sum() function totals 
# them together by column. This is all stored in a variable that is called to realize the output.

num_nan_values = videogames.isna().sum()
print("Number of NaN values in each column:")
print(num_nan_values)
```

## 10. Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
# To calculate the median user score value, we need to import the NumPy library to access its median() function

import numpy as np

# To handle the issue of some of the user score entries being non-numerical string values, I am employing Pandas's 
# to_numeric() function with the errors argument set to coerce. This function and argument specifically targets any values 
# in the User_Score variable that are invalid and turns into NaN values. Then the median() function from NumPy is used on 
# the correctly entered User_Score values to compute the median value and is then stored in a variable. To replace the NaN 
# values in the variable with the newly calculated median value, the fillna() function with the median value as an argument 
# as well as the inplace argument set to True is needed. Fillna() accesses the NaN values and replaces them with the median 
# value to fill the dataset variable. The median user score value is printed and the head() function is set to 15 to see 
# proper implementation of the code.

videogames['User_Score'] = pd.to_numeric(videogames['User_Score'], errors='coerce')
median_user_score = videogames['User_Score'].median()
videogames['User_Score'].fillna(median_user_score, inplace=True)
print('The median user score value is:', median_user_score)
videogames.head(15)
```
