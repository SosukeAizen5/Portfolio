# Program Start

```python
# I will import the dataset (I will be using the Candy dataset for three chapters) and display its contents here. For 
# whatever reason, the dataset was not able to be read by utf-8 encoding, so I used latin1 to read the dataset.

import pandas as pd
import numpy as np

candy = pd.read_csv('candyhierarchy2017.csv', encoding='latin1')

candy.head()
```

## Chapter 7 Transformations

### 1. Filter Out Missing Data

```python
# To filter out missing data, the dropna() function will be used. I will keep the parent dataset intact in case I need to 
# use it later on for another transformation.

candies = candy.dropna()

candies.head()
```

### 2. Fill In Missing Data

```python
# It was very unexpected to see that every single column in the dataset contained missing values, so I'm glad that I kept 
# the parent dataset preserved to make more transformations on that dataset as I would have had nothing left to transform 
# had I not kept it separate. So I will simply continue to use the parent candy variable when needed.

# To fill in missing data properly, I will need to identify the columns with numerical values and those with categorical 
# values and fill them in differently. For the numerical categories, I will fill the missing values with the mean, and the 
# categorical columns will be filled with the most frequently appearing value (the mode). I can find the correct columns by 
# using the select_dtypes() functions and specifying the data types I am looking for.

# To start, I will create a variable that copies the original candy dataset and work with that new variable.

candy_filled = candy

numerical_columns = candy_filled.select_dtypes(include = ['int64', 'float64']).columns

categorical_columns = candy_filled.select_dtypes(include = ['object']).columns

# Now I will need to use fillna() twice: once to fill in the numerical columns with the mean, and once again to fill the 
# categorical columns with the mode.

candy_filled[numerical_columns] = candy_filled[numerical_columns].fillna(candy_filled[numerical_columns].mean())

# For the categorical variables, I used a for loop with the fillna() function to insert the mode using the iloc() function 
# specifying the first index.

for column in categorical_columns:
    candy_filled[column] = candy_filled[column].fillna(candy_filled[column].mode().iloc[0])

candy_filled.head()
```

## Chapter 8 Transformations

### 1. Create Hierarchical Index

```python
# To create a hierarchical index, I chose the second through fifth question columns as they looked to be the easiest to 
# have data be displayed in multiple levels. The set_index() function is what creates the hierarchical index.

hierarchical_columns = ['Q2: GENDER', 'Q3: AGE', 'Q4: COUNTRY', 'Q5: STATE, PROVINCE, COUNTY, ETC']

candy_indexed = candy_filled.set_index(hierarchical_columns)

candy_indexed.head()
```

### 2. Reshape

```python
# To reshape the data, I took the same candy_filled dataset that has been hierarchically indexed and used the stack() 
# function on it. 

stacked_reshaped_candy = candy_indexed.stack()

stacked_reshaped_candy.head()
```

## Chapter 10 Transformations

### 1. Grouping with Index Levels

```python
# Grouping is done with the groupby() function, and to group the hierarchically indexed dataset made in the Chapter 8 
# transformations, I can specify the index level as an argument of groupby(). I chose the first index for convenience 
# (although I did see the other three levels and some of the observations were funny). The book used the count() function 
# also, which I included as well to show the number of observations were in the first index level.

grouped_index_candy = candy_indexed.groupby(level = 0).count()

grouped_index_candy
```

### 2. Split/Apply/Combine

```python
# To perform a split/apply/combine on the data, I will take my already grouped data and apply a function to it with the 
# apply() function. I chose to make a basic sum function to show that when the function is applied to the dataset, the 
# answer across all columns should be the total number of entries. 

def add_it_up(groups):
    return groups.sum()

applied_candy = grouped_index_candy.apply(add_it_up)

applied_candy
```

## Chapter 11 Transformations

### 1. Converting Between String and Datetime

```python
# For this transformation, I need to add a new column to the dataframe containing data that can be converted between string 
# and datetime formats. I will be using the candy_filled variable for ease of reading the dataset. The pd.to_datetime() 
# function is used to convert the string '2020-02-02' to a datetime format in Python. 

# After running the below code, I received a PerformanceWarning that the data was very fragmented. I included and imported 
# the warnings code and module to ignore all warnings so it would not show in the code output.

import warnings

warnings.filterwarnings("ignore")

candy_filled['Date_of_data'] = pd.to_datetime('2020-02-02')

candy_filled.head()
```

```python
# Here, I will convert the newly created datetime column data into string format data using the strftime() function.

candy_filled['Date_of_data'] = candy_filled['Date_of_data'].dt.strftime('%Y-%M-%D')

candy_filled.head()
```

### 2. Generate Date Range

```python
# Now that I have a datetime column (which has now been converted back to string format), I will convert the column back to 
# datetime format, then use the pd.date_range() function to specify a range of dates matching the number of rows within the 
# dataset using the start and periods arguments within the function. 

candy_filled['Date_of_data'] = pd.to_datetime('2020-02-02')

number_of_dates = len(candy_filled)

range_of_dates = pd.date_range(start = '2020-02-02', periods = number_of_dates)

candy_filled['Date_of_data'] = range_of_dates

candy_filled.head()
```
