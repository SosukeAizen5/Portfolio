## Activity 4.01

```python
# 1. Load the necessary libraries. 

# This has already been done, as there will be no need for any additional libraries to be imported. An extra library that I 
# have chosen to import is the warnings library, as further down in my code, warnings are displayed automatically and are 
# not good for reports

import warnings
```

```python
# 2. Read the adult income dataset.

adult_income = pd.read_csv('adult_income_data.csv')
adult_income.head()
```

```python
# 3. Create a script that that will read a text file line by line.

# After downloading and reviewing the text file, I see that the text file contains the column (variable) names needed to 
# label the unlabeled data present in the CSV file. So while having the for loop iterate through each line, I will have the 
# proper column labels split from the text file and placed into a list that I can append to the adult_income data frame. 
# The splitting of the labels will happen on the colon (:) and the index of 0 being called captures everything before the :.

column_names = []
with open('adult_income_names.txt','r') as adult_income_names:
    for line in adult_income_names:
        adult_income_names.readline()
        variable = line.split(":")[0]
        column_names.append(variable)
        
column_names
```

```python
# 4. Add a name of Income for the response variable to the dataset. 

# This is a simple name appending to the column_names list we just created.

column_names.append('Income')

# Now I can append the column_names list to the adult_income data frame to label the currently unlabaled data by using the 
# names argument in the read_csv() function

adult_income = pd.read_csv('adult_income_data.csv', names = column_names)

# The head() function is used here to verify that the labels were appended correctly

adult_income.head()
```

```python
# 5. Find the missing values.

# This is simply done using the isnull() and sum() functions

adult_income.isnull().sum()
```

```python
# 6. Create a DataFrame with only age, education, and occupation by using subsetting.

# To subset the data frame to display one that only holds the above variables, the bracket method can be used

adult_income_subset = adult_income[['age','education','occupation']]

# The head() function is used here to verify that the data frame was created correctly

adult_income_subset.head()
```

```python
# 7. Plot a histogram of age with a bin size of 20.

# The previously used hist() function from Matplotlib is needed for this as well as specifying the bin parameter to be 20. 

plt.hist(adult_income_subset['age'], bins = 20)
plt.title('Histogram of age Variable')
plt.show()
```

```python
# 8. Create a function to strip the whitespace characters.

# To create this function, the strip() function will be built in to the body of the function and the result will be returned

def strip_whitespace(data):
    return data.strip()
```

```python
# 9. Use the apply method to apply this function to all the columns with string values, create a new column, copy the 
# values from this new column to the old column, and drop the new column.

# The only two variables that contain string values are the education and occupation columns

# The resulting code outputs warnings, so I imported the warnings library for this reason and that is to ignore the warning 
# output as it is not professional to output warnings when the code has been implemented

warnings.filterwarnings("ignore")

adult_income_subset['education_no_whitespace'] = adult_income_subset['education'].apply(strip_whitespace)
adult_income_subset['education'] = adult_income_subset['education_no_whitespace']
adult_income_subset.drop(labels = ['education_no_whitespace'],axis=1,inplace=True)

adult_income_subset['occupation_no_whitespace'] = adult_income_subset['occupation'].apply(strip_whitespace)
adult_income_subset['occupation'] = adult_income_subset['occupation_no_whitespace']
adult_income_subset.drop(labels = ['occupation_no_whitespace'],axis=1,inplace=True)

# The head() function is only to verify that the changes have been made properly

adult_income_subset.head()
```

```python
# 10. Find the number of people who are aged between 30 and 50.

# I created two subsets of data using brackets and conditional statements: one subset encompassing those equal or above the 
# age of 30, and the second one subsetting the newly created subset to encompass the people who are 30 or higher, but not 
# over age 50. To realize the number of people who fit those parameters, the len() is used.

adult_age_range_30_and_over = adult_income_subset[(adult_income_subset['age']>=30)]
adult_age_range_30_thru_50 = adult_age_range_30_and_over[(adult_age_range_30_and_over['age']<=50)]
print('The number of people aged 30 through 50 is:', len(adult_age_range_30_thru_50))
```

```python
# 11. Group the records based on age and education to find how the mean age is distributed.

# Peforming this task involves using the groupby() function to group the data by the above variables and then appending the 
# age mean to see its distribution for each age.

mean_age_by_education = adult_income_subset.groupby(['age', 'education'])['age'].mean()
mean_age_by_education
```

```python
# 12. Group by occupation and show the summary statistics of age. Find which profession has the oldest workers on average 
# and which profession has its largest share of the workforce above the 75th percentile.

# For this one, while I can visually figure out the answers needed after the summary statistics are shown, I wanted to find 
# out how I could automate this task provided I were to be given larger datasets that are not so easy to see. I found that 
# Pandas has a function called idxmax(), which returns the index where the first instance of largest value is found. So I 
# will be utlizing this function by storing the summary statistics within a variable and calling the variable to find the 
# values I need.

adult_subset_summary = adult_income_subset.groupby('occupation').describe()['age']
adult_subset_summary
```

```python
oldest_workers = adult_subset_summary['mean'].idxmax()
print('The profession with the oldest workers on average is:', oldest_workers)

share_workforce_above_75 = adult_subset_summary['75%'].idxmax()
print('The profession with the largest share of the workforce above the 75th percentile is:', share_workforce_above_75)
```

```python
# 13. Use subset and groupBy to find the outliers.

# The adult_income_summary variable I created for the previous question contains the outlier values within and used the 
# subset() and groupby() functions to attain them. This variable will be used to answer the next question.
```

```python
# 14. Plot the outlier values on a bar chart. 

# To output the variable the same way that the textbook has done, the barh() function in Matplotlib is needed to realize 
# the necessary horizontal bar graph, with the summary indices being the category parameter and the count observations 

plt.barh(adult_subset_summary.index, adult_subset_summary['count'])
plt.title('Bar Chart of Occupation and Age Summary Outlier Values')
plt.xlabel('Outlier Values')
plt.ylabel('Occupations')
plt.show()
```

```python
# 15. Merge the two DataFrames using common keys to drop duplicate values. 

# Looking at the final output that the textbook wants to be realized, I must create two new subsets and then merge them on 
# a common variable (I will only have one common variable to simplify the merging process) using the inner argument and the 
# drop_duplicates() function.

adult_income_subset_to_merge1 = adult_income[['age', 'workclass', 'occupation']]
adult_income_subset_to_merge2 = adult_income[['education', 'occupation']]

merged_adult_income_subsets = pd.merge(adult_income_subset_to_merge1, adult_income_subset_to_merge2, on='occupation',
                                    how='inner').drop_duplicates()
merged_adult_income_subsets.head()
```

```python
# 1. Add Series 1 and Series 2 together and print the results

# I need to construct the series1 and series2 variables to perform mathemtical operations on them

series1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
series2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])

# Now to add them

# Basic '+' operator

add1 = series1 + series2
print(add1)
```

```python
# Add() function (this function turns indices not present in the series into zeroes with the fill_value argument)

add2 = series1.add(series2, fill_value=0)
print(add2)
```

```python
# Now to subtract them

# Basic '-' operator

subtract1 = series1 - series2
print(subtract1)
```

```python
# Subtract() function

subtract2 = series2.subtract(series1, fill_value=0)
print(subtract2)
```
