# Program Start

```python
# To craft the necessary charts, I will need to import my dataset, which I have chosen to be the Obama Approval Ratings 
# dataset from the Exercise 1.2 Datasets file folder. I have saved that file to a CSV file and will read it in here using 
# Pandas, which will be imported now along with Matplotlib, the library I have chosen to use to output all of the 
# visualizations needed: a bar chart, a stacked bar chart, a pie chart, a donut chart, and a line chart.

import pandas as pd
import matplotlib.pyplot as plt

obama_ratings = pd.read_csv('obama-approval-ratings.csv')

# The head() function is used here to verify that the data has been loaded in correctly.

obama_ratings.head()
```

## Visualization 1: Bar Chart

```python
# For the bar chart, I will be highlighting the percentage of respondents who approve of the issues Obama has addressed in 
# the dataset. Using the barh() function in Matplotlib, I will craft a horizontal bar chart as this chart work better to 
# fully display the names of the political issues within the Issues column. 

plt.barh(obama_ratings['Issue'], obama_ratings['Approve'], color = 'black')
plt.ylabel('Political Issue')
plt.xlabel('Percentage of Approval')
plt.title('Obama Political Issue Approval Percentages')
plt.show()
```

## Visualization 2: Stacked Bar Chart

```python
# In the previous bar chart, I only visualized the approval ratings for these political issues. A stacked bar chart gives 
# the opportunity to plot all three ratings in one chart, one on top of the other. I will use the set_index() function to 
# use each political issue in the Issue column as an index to plot with. Specifying the kind argument as bar and the stacked
# argument as True calls for the crafting of a stacked bar chart. Placing a legend in the customization block of the stacked
# bar chart allows for the viewer to see which bars are meant for which ratings. I toggled with the bbox_to_anchor argument
# in the legend() function so the legend will not be placed on top of the chart but beside it.

obama_ratings.set_index('Issue').plot(kind = 'bar', stacked = True, color = ['black', 'pink', 'yellow'])
plt.xlabel('Political Issue')
plt.ylabel('Rating Percentage')
plt.title('Obama Political Issue Rating Percentages')
plt.legend(bbox_to_anchor = (1.3,0.2))
plt.show()
```

## Visualization 3: Pie Chart

```python
# With the previous stacked bar chart, one could see the proportion of ratings across all political issues. However, it is 
# difficult to discern the percentage of each rating as a whole in this way without the percentage being listed as a number
# as well as these statistics being in bar form along with many other bars. This is where a pie chart can come in handy, as 
# I will create one that shows the proportion of ratings for one individual issue, say the Race Relations issue. Storing 
# the ratings of the Race Relations issue in a variable requires me to locate those ratings using the iloc() indexing 
# function. I then can craft the pie chart using the pie() function with the autopct argument set to %1.1f%% so that the 
# percentage of the whole can be shown on the pie slices.

race_relations = obama_ratings[obama_ratings['Issue'] == 'Race Relations'][['Approve', 'Disapprove', 'None']].iloc[0]

plt.pie(race_relations, labels = ['Approve', 'Disapprove', 'None'], autopct = '%1.1f%%', colors = ['gray', 'yellow', 
                                                                                                   'pink'])
plt.title('Ratings Percentages for the Race Relations Political Issue')
plt.show()
```

## Visualization 4: Donut Chart

```python
# A donut chart is very similar to the pie chart in function and coding, yet an extra argument is added to the pie chart 
# construction to craft a donut chart. I will visualize the Immigration political issue from the Issue column using the same
# methodology as the previous pie chart and add the wedgeprops argument to make the pie chart construction a donut chart.

immigration = obama_ratings[obama_ratings['Issue'] == 'Immigration'][['Approve', 'Disapprove', 'None']].iloc[0]

plt.pie(immigration, labels = ['Approve', 'Disapprove', 'None'], autopct = '%1.1f%%', colors = ['orange', 'blue', 
                                                                                                   'red'], 
        wedgeprops = dict(width = 0.2))
plt.title('Ratings Percentages for the Immigration Political Issue')
plt.show()
```

## Visualization 5: Line Chart

```python
# Line charts are best used for time series data, yet we can also use a line chart here to show the ratings of each 
# political issue. The plot() function is used to plot a dot plot, yet the marker argument alllows the dot plot to be 
# connected by lines to form the line chart. For the issue names being placed on the x-axis, I needed to add in the xticks()
# customization function with the rotation argument set to 90 so the issue names could actually be read, as otherwise they 
# would be overlapping each other, making the x-axis unreadable. I have plotted all three ratings on one visualization here.

plt.plot(obama_ratings['Issue'], obama_ratings['Approve'], marker = 'o', label = 'Approve', color = 'purple')
plt.plot(obama_ratings['Issue'], obama_ratings['Disapprove'], marker = 'o', label = 'Disapprove', color = 'green')
plt.plot(obama_ratings['Issue'], obama_ratings['None'], marker = 'o', label = 'None', color = 'cyan')
plt.xlabel('Political Issue')
plt.ylabel('Rating Percentage')
plt.xticks(rotation = 90)
plt.title('Obama Political Issue Rating Percentages')
plt.legend()
plt.show()
```
