# Program Start

```python
# To craft the necessary charts, I will need to import my dataset, which I have chosen to be the Life Expectancy dataset 
# from the Exercise 4.2 Datasets file folder. I have saved that file to my directory and will read it in here using 
# Pandas, which will be imported now along with Matplotlib and Seaborn, the libraries I have chosen to use to output all 
# of the visualizations needed: a scatter plot, a bubble chart, and a density plot.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

life = pd.read_csv('life-expectancy.csv')

# The head() function is used here to verify that the data has been loaded in correctly.

life.head()
```

## Visualization 1: Scatter Plot

```python
# For the scatter plot, I will construct the chart using the scatter() function to visualize the relationship between the 
# country's life expectancies. As the dataset contains many observations, I will use the head() function to capture only the
# first 30 countries listed. I needed to increase the chart's size using the figure() customization chart function and I 
# also had to adjust the x-axis labels, as the country names needed to be rotated and seen individually as opposed to 
# overlapping one another. This was accomplished with the xticks() function with both the rotation and ha (horizontal 
# alignment) arguments.

plt.figure(figsize = (12,8))
plt.scatter(life['country'].head(30), life['expectancy'].head(30), color = 'blue')
plt.title('Scatter Plot of 2008 Life Expectancy by Country (First 30 Alphabetically)')
plt.xlabel('Country')
plt.ylabel('Life Expectancy in Years')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

## Visualization 2: Bubble Chart

```python
# A bubble chart in Python can be made almost exactly in the same way as a scatter plot is created. I crafted the chart the 
# same way except the s argument and the alpha argument are needed to establish the bubble size and the transparency of the
# bubbles. A bubble chart is different from a scatter plot as it involves a third variable that controls the size of each 
# data point, yet the third variable in this dataset is the year, which is the same across all observations. This means that
# the chart's data points will all be the same size.

plt.figure(figsize=(12,8))
plt.scatter(life['country'].head(30), life['expectancy'].head(30), s = 1000, alpha = 0.5, color = 'green')
plt.title('Bubble Chart of 2008 Life Expectancy by Country (First 30 Alphabetically)')
plt.xlabel('Country')
plt.ylabel('Life Expectancy in Years')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

## Visualization 3: Density Plot

```python
# To craft the density plot, I will be using the kdeplot() function in Seaborn, which is made to craft a kernel density 
# estimate plot. To make sure that the expectancy variable observations are treated as numerical data points, the 
# pd.to_numeric() function has been included. Upon making the plot, I included the fill argument to enhance readability. 

life['expectancy'] = pd.to_numeric(life['expectancy'], errors = 'coerce')

plt.figure(figsize = (10, 6))
sns.kdeplot(life['expectancy'], fill = True, color = 'red')
plt.title("Density Plot of 2008 Life Expectancy by Country (First 30 Alphabetically)")
plt.xlabel("Life Expectancy in Years")
plt.ylabel("Density")
plt.show()
```
