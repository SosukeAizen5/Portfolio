```python
# Exercise 2-4

# Code taken from chap02ex.py file and modified to craft the proper program

from __future__ import print_function

import sys
from operator import itemgetter

import thinkstats2
import first

# Below code has been modified to actually return a mode value instead of 0
# The code will use the max() function on hist.items() and be stored
def Mode(hist):
    
    p, x = max([(p, x) for x, p in hist.Items()])
    return x

# Code below must be modified to actually return sorted histogram values
# reverse parameter must be set to TRUE to achieve the descending order

def AllModes(hist):

    return sorted(hist.Items(), key=itemgetter(1), reverse=True)

# New function needs to be defined to successfully output a quantifiable
# difference of the weights of first babies and other babies
# The two statistics needed are what the book mentions: mean and variance


def WeightDifference(live, firsts, others):
    
    # mean is taken from the totalwgt_lb variable in both datasets
    
    mean1 = firsts.totalwgt_lb.mean()
    mean2 = others.totalwgt_lb.mean()
    
    # variance is taken from the totalwgt_lb variable in both datasets
    
    var1 = firsts.totalwgt_lb.var()
    var2 = others.totalwgt_lb.var()
    
    # Print statements are needed to display these figures in a reabable
    # format for users to understand the statistics
    
    print('Mean')
    print('First babies', mean1)
    print('Others', mean2)

    print('Variance')
    print('First babies', var1)
    print('Others', var2)
    
    # Print statements for the actual difference in weight in pounds and ounces
    # are needed by subtracting the means from each other and multiplying by 16
    # for the difference in ounces number
    
    print('Difference in lbs', mean1 - mean2)
    print('Difference in oz', (mean1 - mean2) * 16)
    
    # Code and appropriate print statement is needed for the computation of 
    # Cohen's d to measure effect size
    
    cohen_d = thinkstats2.CohenEffectSize(firsts.totalwgt_lb, others.totalwgt_lb)
    print("Cohen's d", cohen_d)

# Code must be altered to correct errors in coding and display all statistics


def main():

    live, firsts, others = first.MakeFrames()
    hist = thinkstats2.Hist(live.prglngth)
    
    # WeightDifference() function must be called in main()
    
    WeightDifference(live, firsts, others)

    # test Mode    
    mode = Mode(hist)
    print('Mode of pregnancy length', mode)
    assert(mode == 39)

    # test AllModes
    modes = AllModes(hist)
    assert(modes[0][1] == 4693)

    for value, freq in modes[:5]:
        print(value, freq)


if __name__ == '__main__':
    main()
```
