```python
# Summary statistics of each variable

# In order to calculate the mean, mode, spread, and tails of the variables, I will be using the corresponding thinkstats2 
# functions: Mean() for the mean, Mode() for the mode, Var() and Std() for spread, and Skewness() and 
# PearsonMedianSkewness() for the tails indicating skewness and/or kurtosis

year_mean = thinkstats2.Mean(immigration[year])
year_mode = immigration[year].mode()
year_var = thinkstats2.Var(immigration[year])
year_std = thinkstats2.Std(immigration[year])
year_skew = thinkstats2.Skewness(immigration[year])
year_kurt = thinkstats2.PearsonMedianSkewness(immigration[year])

print('Year variable')
print("Mean:", year_mean)
print("Mode:", year_mode)
print("Variance:", year_var)
print("Standard Deviation:", year_std)
print("Skewness:", year_skew)
print("Kurtosis:", year_kurt)
```

```python
# PMF scenarios

# To accopmplish outlining two scenarios with the same variable PMF, I will need to split the dataset in two. I have chosen
# to split the dataset by the year variable to make the comparison of immigrants receiving their permanent residences within
# a certain range of years versus all other years

split_year1 = 2010

immigration_before_2010 = immigration[immigration[year] < split_year1]
immigration_2010_and_beyond = immigration[immigration[year] >= split_year1]

before_2010_pmf = thinkstats2.Pmf(immigration_before_2010[perm_residence], 
                                  label='Immigrant Permanent Residences Before 2010')
at_and_after_2010_pmf = thinkstats2.Pmf(immigration_2010_and_beyond[perm_residence], 
                                        label='Immigrant Permanent Residences 2010 and After')

thinkplot.PrePlot(2)
thinkplot.Pmf(before_2010_pmf, color='blue')
thinkplot.Pmf(at_and_after_2010_pmf, color='red')
thinkplot.Show(xlabel='Value', ylabel='Probability')
```

```python
# CDF of perm_residence variable

perm_res_cdf = thinkstats2.Cdf(immigration[perm_residence])

# Plot the CDF
thinkplot.Cdf(perm_res_cdf)
thinkplot.Show(xlabel='Value', ylabel='CDF')
```

```python
# Pareto distribution for the perm_residence variable

shape, loc, scale = scipy.stats.pareto.fit(immigration[perm_residence])

pareto_dist = scipy.stats.pareto(shape, loc, scale)

x = np.linspace(pareto_dist.ppf(0.01), pareto_dist.ppf(0.99), 100)
pdf = pareto_dist.pdf(x)

plt.plot(x, pdf, 'b-', lw=2)

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Pareto Distribution of perm_residence Variable')
plt.show()
```

```python
# Correlation test of perm_residence and refugee variables

# Below function is taken from the Thinkstats Exercise file on Chapter 9

class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys

hypo_test_data = immigration[perm_residence], immigration[refugee]
corr_test = CorrelationPermute(hypo_test_data)
pvalue = corr_test.PValue()
pvalue
```

```python
# Comparing the absolute value of the actual correlation to the largest realized value in the simulations

corr_test.actual, corr_test.MaxTestStat()
```
